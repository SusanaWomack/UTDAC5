```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Reading Data Set using Pandas

```python
credit = pd.read_csv("default of credit card clients.csv", header = 1)
```

## Data Preprocessing

```python
# rename columns
credit.rename(columns = {'default payment next month': 'DEFAULT_IND'}, inplace=True)
credit.rename(columns = {'PAY_0': 'PAY_1'}, inplace=True)
```

```python
# discretize age colummn
credit['AGE_BIN'] = pd.cut(credit['AGE'], bins=[17,25,35,45,55,67,100])
```

```python
# count clients in new age bins
credit.groupby('AGE_BIN')['AGE_BIN'].count()
```

```python
# check for nulls
credit[credit['AGE_BIN'].isnull() == True]
```

```python
# Number of clients in each education bucket
credit.groupby('EDUCATION')['EDUCATION'].count()
```

```python
# group all "other" education types (0,4,5,6) into one value (0)
credit.loc[credit['EDUCATION'].isin([4,5,6]), 'EDUCATION'] = 0
```

```python
# Verify update and review number of clients in each education bucket
credit.groupby('EDUCATION')['EDUCATION'].count()
```
